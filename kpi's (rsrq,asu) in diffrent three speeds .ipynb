```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm  # For color mapping
import numpy as np  # For color mapping

def plot_rsrq_all_files(directory_path):
    # Check if the directory path is valid
    if not os.path.isdir(directory_path):
        print(f"Invalid directory path: {directory_path}")
        return

    # Log to track which files are processed
    print("Starting to process files...")

    # Create a single plot
    plt.figure(figsize=(12, 8))

    # Generate a list of colors using a color map
    color_map = cm.rainbow(np.linspace(0, 1, len(os.listdir(directory_path))))

    # Iterate over each CSV file in the directory
    for index, filename in enumerate(os.listdir(directory_path)):
        # Check if the file has a CSV extension (case-insensitive)
        if filename.lower().endswith('.csv'):
            file_path = os.path.join(directory_path, filename)
            print(f"Processing file: {filename}")

            try:
                # Read the CSV file
                df = pd.read_csv(file_path)

                # Check expected columns
                expected_columns = ['  currentTime', ' rsrq', ' speed', ' asu']
                if not all(col in df.columns for col in expected_columns):
                    print(f"Skipping file {filename}: Expected columns not found.")
                    continue

                # Convert 'currentTime' to datetime
                df['  currentTime'] = pd.to_datetime(df['  currentTime'])
                df['Time_Seconds'] = (df['  currentTime'] - df['  currentTime'].min()).dt.total_seconds()

                # Filter data to only include within the first 50 seconds
                df = df[df['Time_Seconds'] <= 50]

                # Plot RSRQ vs Time for the current file
                plt.scatter(df['Time_Seconds'], df[' rsrq'], color=color_map[index], alpha=0.5, label=filename,s=10)

            except Exception as e:
                print(f"Error processing file {filename}: {e}")

        else:
            print(f"Skipping non-CSV file: {filename}")

    # Configure the plot
    plt.xlabel('Time (seconds)')
    plt.ylabel('RSRQ (dBm)')
    plt.title('RSRQ vs Time (0-50 seconds) for a three different speeds ')
    plt.grid(True)
    
    plt.tight_layout()

    # Display the plot
    plt.show()

# Set the directory path and call the function
directory_path = 'C:/Users/ASUS/Desktop/pfe memoire/spee'
plot_rsrq_all_files(directory_path)
```
