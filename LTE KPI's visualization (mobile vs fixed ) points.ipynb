```python
import os
import pandas as pd
import matplotlib.pyplot as plt


# Function to plot fixed and mobile points within a specified time range (0 to 300 seconds)
def plot_points_from_directories(directory_path):
    # Initialize empty lists to store fixed and mobile data
    fixed_points_all = []
    mobile_points_all = []

    # Iterate over each CSV file in the directory
    for filename in os.listdir(directory_path):
        if filename.endswith('.csv'):
            file_path = os.path.join(directory_path, filename)

            try:
                # Read the CSV file
                df = pd.read_csv(file_path)

                # Check expected columns exist (considering extra leading spaces)
                expected_columns = ['  currentTime', ' rsrp', ' speed', ' asu']
                if not all(col in df.columns for col in expected_columns):
                    print(f"Skipping file {filename}: Expected columns not found.")
                    continue

                # Convert 'currentTime' to datetime and rebase to 0 seconds
                df['  currentTime'] = pd.to_datetime(df['  currentTime'])
                df['Time_Seconds'] = (df['  currentTime'] - df['  currentTime'].min()).dt.total_seconds()

                # Filter data to include only between 0 and 300 seconds
                df = df[(df['Time_Seconds'] >= 0) & (df['Time_Seconds'] <= 300)]

                # Separate fixed and mobile points based on speed
                fixed_points = df[
                    (df[' speed'] < 1) & (df[' asu'] >= 15) & (df[' asu'] <= 90)
                ]
                mobile_points = df[
                    (df[' speed'] >= 1) & (df[' asu'] >= 15) & (df[' asu'] <= 90)
                ]

                # Store fixed and mobile data for later plotting
                fixed_points_all.append(fixed_points)
                mobile_points_all.append(mobile_points)

            except Exception as e:
                print(f"Error processing file {filename}: {e}")

    # Create a plot to display the results
    plt.figure(figsize=(10, 6))

    # Plot mobile points as red scatter plots
    for mobile_points in mobile_points_all:
        plt.plot(
            mobile_points['Time_Seconds'],
            mobile_points[' rsrq'],
            color='red',
            alpha=0.6,
            label='Mobile Points',
        )

    # Plot fixed points as black lines
    for fixed_points in fixed_points_all:
        plt.plot(
            fixed_points['Time_Seconds'],
            fixed_points[' rsrq'],
            color='blue',
            alpha=0.8,
            label='Fixed Points',
        )

    # Set plot labels and title
    plt.xlabel('Time (seconds)')
    plt.ylabel('RSRQ (dBm)')
    plt.title('RSRQ vs Time (0 - 300 seconds) 4:00 PM Batna')
    plt.grid(True)
    plt.xticks(rotation=45)
    
    # Add a legend to the plot
    #plt.legend()

    # Adjust layout and display the plot
    plt.tight_layout()
    plt.show()


# Set the directory path and call the function
directory_path = 'C:/Users/ASUS/Desktop/pfe memoire/new'
plot_points_from_directories(directory_path)
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

def plot_points_from_directories(directory_path):
    # Initialize empty list to store RSRP values
    rsrp_values = []

    # Iterate over each file in the directory
    for filename in os.listdir(directory_path):
        if filename.endswith('.csv'):
            file_path = os.path.join(directory_path, filename)
            
            try:
                # Read the CSV file
                df = pd.read_csv(file_path)
                
                # Check if 'currentTime' column exists
                if '  currentTime' not in df.columns:
                    print(f"Skipping file {filename}: 'currentTime' column not found.")
                    continue
                
                # Convert 'currentTime' column to datetime type
                df['  currentTime'] = pd.to_datetime(df['  currentTime'])
                
                # Extract time component and convert to minutes
                df['time_minutes'] = (df['  currentTime'] - df['  currentTime'].min()).dt.total_seconds() / 60
                
                # Filter data for the time range from 0 to 5 minutes
                df_filtered = df[(df['time_minutes'] >= 0) & (df['time_minutes'] <= 5)]
                
                # Append RSRP values to the list
                rsrp_values.extend(df_filtered[' rsrp'].tolist())
                
            except Exception as e:
                print(f"Error processing file {filename}: {e}")

    # Plotting
    plt.figure(figsize=(10, 6))

    # Plot RSRP values
    plt.plot(rsrp_values, color='blue', alpha=0.7)
    
    plt.xlabel('Time (seconds)')
    plt.ylabel('RSRP')
    plt.title('RSRP Variation over Time')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

directory_path = 'C:/Users/ASUS/Desktop/pfe memoire/graphs/sou'
plot_points_from_directories(directory_path)
```
